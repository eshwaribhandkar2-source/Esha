```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
os.environ["OMP_NUM_THREADS"]="1"
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df.head()
df.tail()
df.describe()
df.info()
df.isnull().sum()
```

```python
df.shape
```

```python
x=df.iloc[:,[1,2,3,4]].values
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
wcss=[]
for i in range(1,11):
    Kmeans=KMeans(n_clusters=i, init= 'k-means++', random_state=42)
    Kmeans.fit(x_scaled)
    wcss.append(Kmeans.inertia_)

plt.plot(range(1,11),wcss,marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("wcss")
plt.show()
```

```python
 Kmeans=KMeans(n_clusters=3, init= 'k-means++', random_state=42)
y_pred= Kmeans.fit_predict(x_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_scaled[y_pred==0,0],x_scaled[y_pred==0,1],s=100,c="red",label="Cluster 1")
plt.scatter(x_scaled[y_pred==1,0],x_scaled[y_pred==1,1],s=100,c="green",label="Cluster 2")
plt.scatter(x_scaled[y_pred==2,0],x_scaled[y_pred==2,1],s=100,c="blue",label="Cluster 3")
plt.scatter(Kmeans.cluster_centers_[:,0], Kmeans.cluster_centers_[:,1],
            s=200,c="yellow",label="Centroids")
plt.title("iris")
plt.legend()
plt.show()

                      
```
